Task - 1:
Create a Python program that converts between different units of measurement.
• The program should:
1. Prompt the user to choose the type of conversion (e.g., length, weight, volume).
2. Ask the user to input the value to be converted.
3. Perform the conversion and display the result.
4. Handle potential errors, such as invalid input or unsupported conversion types.
• Requirements:
1. Functions: Define at least one function to perform the conversion.
2. Error Handling: Use try-except blocks to handle invalid input (e.g., non-numeric values).
3. User Input: Prompt the user to select the conversion type and input the value.
4. Docstrings: Include a docstring in your function to describe its purpose, parameters, and
return value.
• Conversion Options:
1. Length:
– Convert meters (m) to feet (ft).
– Convert feet (ft) to meters (m).
2. Weight:
– Convert kilograms (kg) to pounds (lbs).
– Convert pounds (lbs) to kilograms (kg).
3. Volume:
– Convert liters (L) to gallons (gal).
– Convert gallons (gal) to liters (L).

```python
def convert_units(value, conversion_type):
    """
    Performs the mathematical conversion between different units.

    Parameters:
    value (float): The numeric value to be converted.
    conversion_type (str): The specific conversion mapping (e.g., 'm_to_ft').

    Returns:
    float: The converted value.
    str: The unit label for the result.
    """
    # Conversion Constants
    M_TO_FT = 3.28084
    KG_TO_LBS = 2.20462
    L_TO_GAL = 0.264172

    if conversion_type == '1': # Meters to Feet
        return value * M_TO_FT, "feet (ft)"
    elif conversion_type == '2': # Feet to Meters
        return value / M_TO_FT, "meters (m)"
    elif conversion_type == '3': # KG to LBS
        return value * KG_TO_LBS, "pounds (lbs)"
    elif conversion_type == '4': # LBS to KG
        return value / KG_TO_LBS, "kilograms (kg)"
    elif conversion_type == '5': # Liters to Gallons
        return value * L_TO_GAL, "gallons (gal)"
    elif conversion_type == '6': # Gallons to Liters
        return value / L_TO_GAL, "liters (L)"
    else:
        raise ValueError("Unsupported conversion type.")

def main():
    print("--- Professional Unit Converter ---")
    print("Select a conversion:")
    print("1: Meters to Feet  | 2: Feet to Meters")
    print("3: KG to Pounds    | 4: Pounds to KG")
    print("5: Liters to Gal   | 6: Gal to Liters")

    try:
        choice = input("\nEnter choice (1-6): ")

        if choice not in ['1', '2', '3', '4', '5', '6']:
            print("Error: Invalid selection. Please choose 1-6.")
            return

        val_input = input("Enter the value to convert: ")
        val = float(val_input)

        result, unit = convert_units(val, choice)
        print(f"\nResult: {val} converts to {result:.2f} {unit}")

    except ValueError:
        print("Error: Please enter a valid numeric value.")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")

if __name__ == "__main__":
    main()
```

1. Flatten a Nested List:
Write a Python function that takes a nested list and flattens it into a single list, where all the elements
are in a single dimension.
• Requirements:
– Define a function flatten(lst) that takes a nested list lst and returns a flattened version
of the list.
– Example: For the input [[1, 2], [3, 4], [5]], the output should be [1, 2, 3, 4, 5].

```python
def flatten(lst):
    """
    Converts a 2D nested list into a 1D flat list.

    Parameters:
    lst (list): A list containing other lists.

    Returns:
    list: A single list containing all individual elements.
    """
    return [item for sublist in lst for item in sublist]

nested_input = [[1, 2], [3, 4], [5]]
flat_output = flatten(nested_input)

print(f"Original Nested List: {nested_input}")
print(f"Flattened List:       {flat_output}")
```

2. Accessing Nested List Elements:
Write a Python function that extracts a specific element from a nested list given its indices.
• Requirements:
– Define a function access nested element(lst, indices) that takes a nested list lst and
a list of indices indices, and returns the element at that position.
– Example: For the input lst = [[1, 2, 3], [4, 5, 6], [7, 8, 9]] with indices = [1,
2], the output should be 6.

```python
def access_nested_element(lst, indices):
    """
    Extracts an element from a nested list using a list of indices.

    Parameters:
    lst (list): The nested list (could be 2D, 3D, etc.).
    indices (list): A list of integers representing the path to the element.

    Returns:
    The element found at the specified indices.
    """
    current_level = lst
    try:
        for index in indices:
            current_level = current_level[index]
        return current_level
    except (IndexError, TypeError):
        return "Error: Invalid indices for the provided nested list."

matrix = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
target_indices = [1, 2]

result = access_nested_element(matrix, target_indices)

print(f"Matrix: {matrix}")
print(f"Indices: {target_indices}")
print(f"Element Found: {result}")
```

3. Sum of All Elements in a Nested List:
Write a Python function that calculates the sum of all the numbers in a nested list (regardless of depth).
• Requirements:
– Define a function sum nested(lst) that takes a nested list lst and returns the sum of all
the elements.
– Example: For the input [[1, 2], [3, [4, 5]], 6], the output should be 21.

```python
def sum_nested(lst):
    """
    Calculates the sum of all numbers in a list, regardless of nesting depth.

    Parameters:
    lst (list): A list containing numbers or further nested lists.

    Returns:
    int/float: The total sum of all numeric elements.
    """
    total = 0
    for item in lst:
        if isinstance(item, list):
            total += sum_nested(item)
        else:
            total += item
    return total

nested_data = [[1, 2], [3, [4, 5]], 6]
result = sum_nested(nested_data)

print(f"Nested List: {nested_data}")
print(f"Total Sum:   {result}")
```

5. Find the Maximum Element in a Nested List:
Write a Python function that finds the maximum element in a nested list (regardless of depth).
• Requirements:
– Define a function find max(lst) that takes a nested list lst and returns the maximum
element.
– Example: For the input [[1, 2], [3, [4, 5]], 6], the output should be 6.

```python
def find_max(lst):
    """
    Finds the maximum numeric value in a nested list of any depth.

    Parameters:
    lst (list): A nested list containing numbers.

    Returns:
    float/int: The largest number found within the structure.
    """
    max_val = float('-inf')

    for item in lst:
        if isinstance(item, list):
            current_max = find_max(item)
        else:
            current_max = item

        if current_max > max_val:
            max_val = current_max

    return max_val

nested_input = [[1, 2], [3, [4, 5]], 6]
result = find_max(nested_input)

print(f"Nested List: {nested_input}")
print(f"Maximum Element: {result}")
```

6. Count Occurrences of an Element in a Nested List:
Write a Python function that counts how many times a specific element appears in a nested list.
• Requirements:
– Define a function count occurrences(lst, elem) that counts the occurrences of elem in
the nested list lst.
– Example: For the input lst = [[1, 2], [2, 3], [2, 4]] and elem = 2, the output should
be 3.

```python
def count_occurrences(lst, elem):
    """
    Counts the total number of times a specific element appears in a nested list.

    Parameters:
    lst (list): The nested list to search.
    elem: The element to count.

    Returns:
    int: The total count of occurrences.
    """
    count = 0
    for item in lst:
        if isinstance(item, list):
            count += count_occurrences(item, elem)
        else:
            if item == elem:
                count += 1
    return count

nested_list = [[1, 2], [2, 3], [2, 4]]
target_element = 2

total_count = count_occurrences(nested_list, target_element)

print(f"Nested List: {nested_list}")
print(f"Occurrences of '{target_element}': {total_count}")
```

7. Flatten a List of Lists of Lists:
Write a Python function that flattens a list of lists of lists into a single list, regardless of the depth.
• Requirements:
– Define a function deep flatten(lst) that takes a deeply nested list lst and returns a single
flattened list.
– Example: For the input [[[1, 2], [3, 4]], [[5, 6], [7, 8]]], the output should be
[1, 2, 3, 4, 5, 6, 7, 8].

```python
def deep_flatten(lst):
    """
    Recursively flattens a list of any depth into a single dimension.

    Parameters:
    lst (list): A deeply nested list (e.g., list of lists of lists).

    Returns:
    list: A single-dimensional list containing all elements.
    """
    flat_list = []

    for item in lst:
        if isinstance(item, list):
            flat_list.extend(deep_flatten(item))
        else:
            flat_list.append(item)

    return flat_list

deep_nested_input = [[[1, 2], [3, 4]], [[5, 6], [7, 8]]]
result = deep_flatten(deep_nested_input)

print(f"Deeply Nested: {deep_nested_input}")
print(f"Flattened:     {result}")
```

8. Nested List Average:
Write a Python function that calculates the average of all elements in a nested list.
• Requirements:
– Define a function average nested(lst) that takes a nested list lst and returns the average
of all the elements.
– Example: For the input [[1, 2], [3, 4], [5, 6]], the output should be 3.5.

```python
def sum_nested(lst):
    """Recursively calculates the total sum of all numbers."""
    total = 0
    for item in lst:
        if isinstance(item, list):
            total += sum_nested(item)
        else:
            total += item
    return total

def count_nested(lst):
    """Recursively counts the total number of numeric elements."""
    count = 0
    for item in lst:
        if isinstance(item, list):
            count += count_nested(item)
        else:
            count += 1
    return count

def average_nested(lst):
    """
    Calculates the average of all elements in a nested list.

    Parameters:
    lst (list): The nested list containing numbers.

    Returns:
    float: The average of all numbers.
    """
    total_sum = sum_nested(lst)
    total_count = count_nested(lst)

    if total_count == 0:
        return 0

    return total_sum / total_count

nested_input = [[1, 2], [3, 4], [5, 6]]
result = average_nested(nested_input)

print(f"Nested List: {nested_input}")
print(f"Average:     {result}")
```
